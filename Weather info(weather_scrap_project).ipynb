```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.timeanddate.com/weather/india/new-delhi/ext"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('table',id='wt-ext')
table=table.find('tbody')
table
```

```python
rows = table.find_all("tr")
rows
```

```python
rows = table.find_all("tr",style="")
rows
```

```python
Column_info = []

for i in rows:
    cols = i.find_all("td")
    Weather_info = [P.text for P in cols]
    Column_info.append(Weather_info)
print(Column_info)
```

```python
Day = []
Temperature = []
Weather = []
Feels_likes = []
Wind = []
Humidity = []
Chance = []
Amount = []
UV = []
Sunrise = []
Sunset = []
for i in Column_info:
    Day.append(i[0])
    Temperature.append(i[1])
    Weather.append(i[2])
    Feels_likes.append(i[3])
    Wind.append(i[4])
    Humidity.append(i[5])
    Chance.append(i[6])
    Amount.append(i[7])
    UV.append(i[8])
    Sunrise.append(i[9])
    Sunset.append(i[10])
```

```python
Data={"Day":Day,"Temperature":Temperature,"Weather":Weather,"Feels_likes":Feels_likes,"Wind":Wind,"Humidity":Humidity,"Chance":Chance,"Amount":Amount,"UV":UV,"Sunrise":Sunrise,"Sunset":Sunset}
```

```python
weather_info = pd.DataFrame(Data)
weather_info
```

```python
weather_info.to_csv("Weather Info.csv")
```
